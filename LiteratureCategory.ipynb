```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
url = "https://arbuz.kz/ru/almaty/catalog/cat/19338-ovoshi_zelen_griby_solenya#/"
request = requests.get(url)
```

```python
#Problem 2
bs4 = BeautifulSoup(request.text, "html.parser")

for block in bs4.find_all('div', attrs={'class': 'row products-list'}):
    for vegetable_name in block.find_all('div', attrs={'class': 'product-item-text'}):
           print(vegetable_name.get_text())

        
for block in bs4.find_all('div', attrs={'class': 'row products-list'}):
    for price in block.find_all('div', attrs={'class': 'mr-2'}):
            print(price.get_text())
```

```python
#problem 3
from bs4 import BeautifulSoup as bs

titles = []
prices = []

for t in range(84):
    url = "https://www.meloman.kz/fiction/fiction-and-fantasy/?p={}".format(t)
    r = requests.get(url)
    soup  = bs(r.content, 'html.parser')

    titles += [i.text for i in soup .select('.product-item-link')]
    prices += [i.text for i in soup .select('.price')]

df = pd.DataFrame(list(zip(titles, prices)), columns = ['Title', 'Price'])
# df.to_csv(r'Data.csv', sep=',', encoding='utf-8-sig',index = False )
```

```python

df['Price'] = df['Price'].str.strip('₸')
df['Price'] = df['Price'].astype(str).str.replace('\s+', '').replace('nan', np.nan)
df['Price'] = pd.to_numeric(df['Price'])

df['Price']
```
