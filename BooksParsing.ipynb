```python
#Parsing start
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.meloman.kz/books/fiction/'
page = requests.get(url)
print(page.status_code)
```

```python
soup = BeautifulSoup(page.text, "html.parser")
print(soup)
```

```python

book1 = soup.find('span', class_= 'product-item__stock-label')
```

```python
book1.text
```

```python
book_names = []
books_availability = []
for block in soup.findAll('div', class_= 'product details product-item-details'):
    name_str = block.find('a', class_='product-item-link').text
    availability_str = block.find('span', class_= 'product-item__stock-label').text
    book_names.append(name_str)
    books_availability.append(availability_str)
```

```python
book_names
```

```python
books_availability
```

```python
data = {
    'Books': book_names,
    'Availabilty': books_availability
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.to_csv('meloman_books_and_availability.csv', index=False)
```

```python
bookLinks = []
for link in soup.findAll('li', class_='item product product-item'):
    book_link = link.find('a',class_='product photo product-item-photo')['href']
    bookLinks.append(book_link)
    print(bookLinks)
    
```

```python
book_parameters = []
book_info = []
books = []
for link in bookLinks:
    page = requests.get(link)
    soup = BeautifulSoup(page.text, "html.parser")
    table = soup.find('table', class_= 'data table additional-attributes')
    d = {}
    for item in soup.findAll('div', class_= 'column main'):
        book_info_dict = {
        'Book name': item.find('span', class_= 'base').text,
        'Book_price': item.find('span', class_= 'price').text
        }
    book_info.append(book_info_dict)  
    #print(book_info)
    for parameter in table.findAll('tr'):
        d[parameter.find('th').text] = parameter.find('td').text
    book_parameters.append(d)
    #print(book_parameters)
```

```python
notConnectBook = pd.DataFrame.from_dict(book_parameters+book_info)
```

```python
notConnectBook.to_csv('NotConnectBooks.csv', index = True)
```
